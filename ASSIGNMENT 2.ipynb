## 4) Top 5 cars based on sales price

```python
df=pd.DataFrame(Data)
```

```python
top5sales = df.sort_values("Sale Price",ascending=False)
```

```python
top5sales
```

```python
x = [top5sales.iloc[0:5]['Make'],top5sales.iloc[0:5]['Model']]
```

```python
y=pd.DataFrame(x)
```

```python
y
```

```python
#Taking transpose of y
z=y.T
```

```python
z
```

```python
print('Make and Model of top 5 cars based on sale prices is:')
print(z)
```
